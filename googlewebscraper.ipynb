```python
# Wait for the page to load
time.sleep(2)

# Find image elements
imageposts = driver.find_elements(By.CLASS_NAME, "YQ4gaf")
```

```python
for image in imageposts:
    print(image.get_attribute('src'))
```

```python
import base64
import numpy as np


os.makedirs(output_dir, exist_ok=True)
for image in imageposts:
    bruh = image.get_attribute('src')
    if bruh[:4] == 'data' and 'gif' not in bruh.split('/')[1]:

        # Base64 image string
        base64_image = bruh
        # Remove the 'data:image/jpeg;base64,' prefix
        image_data = base64.b64decode(base64_image.split(',')[1])

        # Write the image to a file
        with open(f'screenshots/{np.random.randint(10000,99999)}.jpg', 'wb') as f:
            f.write(image_data)
```
